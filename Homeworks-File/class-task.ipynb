```python
sal = pd.read_csv('Salaries.csv')
```

```python
# 3. Check head of DataFrame
print("Head of DataFrame:")
print(sal.head())
```

```python
# 4. Check and handle null values
print("Null values in dataset:")
print(sal.isnull().sum())
sal = sal.drop(['Notes', 'Status'], axis=1)
print("As the Notes and Status columns are not required for analysis, they have been dropped.")
salDropNull=sal.dropna()
print("Null values in dataset after dropping:")
print(salDropNull.isnull().sum())
```

```python
# 5. Create plots
sns.pairplot(sal)
plt.show()

sns.histplot(sal['BasePay'],kde=True)
plt.title("Distribution of Salaries")
plt.show()

plt.figure(figsize=(10,8))
numerical_data = sal.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numerical_data.corr(), annot=True, cmap="coolwarm",fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
