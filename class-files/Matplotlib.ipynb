```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(0, 5)
y = x ** 2
```

```python
plt.plot(x, y, 'r')
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

```python
plt.subplot(2,3,1)
plt.plot(x, y, 'r--')
plt.subplot(2,3,2)
plt.plot(y, x, 'g*-')
plt.subplot(2,3,3)
plt.plot(y, x, 'r-')
plt.subplot(2,3,4)
plt.plot(y, x, 'r-')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.3, 0.5, 0.7])
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```
